```python
import os
def main():
    # Define the headers for the CSV file
    headers = ["Review ID", "Review Content", "Review Score", "Review Time"]
    
    # Directory to save CSV files
    output_dir = "tripadvisor_hotel_reviews"
    os.makedirs(output_dir, exist_ok=True)

    # List of hotels with names and IDs
    hotels = [
        {"name": "Waterfront Hotel and Casino", "id": 308375},
        {"name": "Seda Ayala Center Cebu", "id": 15129214},
        {"name": "bai Hotel", "id": 11687524},
        {"name": "Fairfield by Marriott Cebu", "id": 27098289},
        {"name": "Jpark Island Resort and Waterpark", "id": 1444058},
        {"name": "Dusit Thani Mactan", "id": 17334188}
    ]
      
    limit = 20

    for hotel in hotels:
        offset = 0
        hotel_name = hotel["name"]
        hotel_id = hotel["id"]
        total_reviews_written = 0
        csv_file = os.path.join(output_dir, f"{hotel_name.replace(' ', '_').lower()}_reviews_2022_2024.csv")
        
        check_and_write_headers(csv_file, headers)
        
        print(f"Scraping {hotel_name}")
        while True:
            # Fetch reviews
            response_data = get_hotel_reviews(hotel_id, limit, offset)
            reviews = response_data[0].get('data',{}).get('HPS_getDefaultReviewsPage',{}).get('reviewListPage',{}).get('reviews',[])
            
            if not reviews:
                print('no reviews left')
                break
            
            # Filter reviews
            filtered_reviews,  old_review = filter_reviews(reviews)
            
            # Write reviews to CSV and get the count of rows written
            row_count = write_reviews_to_csv(csv_file, filtered_reviews)
            
            total_reviews_written += row_count
            offset += limit
            
            if old_review:
                break
            
        print(f"offset {offset}")
        print(f"English reviews from 2022 to 2024 for hotel {hotel_name} have been written to {csv_file}")
        print(f"Total number of rows written for hotel {hotel_name}: {total_reviews_written}\n")
        

if __name__ == "__main__":
    main()
```
